```python

from datetime import datetime
from pyspark.sql import functions as F

today = datetime.today()
first_day = today.replace(day=1)
month_reference = first_day.strftime("%Y-%m-%d")

s3_path = f"s3://clinvar-monthly-342016537087-us-east-2-an/raw/{month_reference}/ClinVarVCVRelease_{month_reference}.xml.gz"

df_raw = spark.read \
    .format("xml") \
    .option("rowTag", "VariationArchive")\
    .option("excludeAttribute", "false")\
    .option("valueTag", "_VALUE")\
    .option("samplingRatio", 0.1) \
    .load(s3_path)\
    .withColumn("ingestion_date", F.current_date())
    
df_raw.write.format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze_clinvar.bronze_clinvar_raw")
```
